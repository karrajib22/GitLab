```python
'''
creatig a new dataframe where all regions will be a column each, 
Date will be index and rows will be values of price25ma
'''

graph_df = pd.DataFrame()

for i in df['region'].unique():
    print(i)
    region_df = df.copy()[df['region']==i]
    region_df.sort_index(inplace=True)
    region_df[f"{i}_price25ma"]=region_df["AveragePrice"].rolling(25).mean()
    
#     graph_df = graph_df.join(region_df[f"{i}_price25ma"])
    
    if graph_df.empty:
        graph_df = region_df[[f"{i}_price25ma"]]
    else:
        graph_df = graph_df.join(region_df[f"{i}_price25ma"])
    
```
